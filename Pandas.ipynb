## **Pandas Series**
- A Pandas Series is similar to a table's column, functioning as a one-dimensional array that can store data of any type.

````python
x = [1, 2, 3]
nums = pd.Series(x)
print(nums)
````

## **Labels**
- If no details are given, each value is marked with its index number, starting from 0 for the first value, 1 for the second, and so on. This index can be used to find a specific value.

- You can create custom labels using the `index` argument.

When you have created labels, you can access an item by referring to the label.

````python
for i in range(len(nums)):
  print(nums[i])

print("----To create Custom Labels-----")
nums = pd.Series(x, index = ['a','b', 'c'])
print(nums)
print(nums["b"])
````

### **Key/Value Objects as Series**
- You can create a Series using a key/value object, similar to a dictionary.

Use the dictionary keys as labels, and if you want specific items, use the `index` argument to select only those items for the Series.

````python
items = {"fruits": 'orange', "vegetables": 'spinach', "N": 45}
var = pd.Series(items)
print(var)
print("---------------------------------")
nums = pd.Series(items, index = ["fruits", "vegetables"])
print(nums)
````

# **DataFrames**
- In Pandas, data is organized into tables called DataFrames, where each column is like a Series, and the entire table is the DataFrame.
- A Pandas DataFrame is like a table with rows and columns, forming a 2D data structure similar to an array.




````python
data = {"Number of Steps": [2000, 5000, 10000],
        "Duration": [25, 30, 70],
        "Calories-burnt": [100, 200, 350]}

table = pd.DataFrame(data)
print(table)
````

You can use Pandas' `loc` attribute to get one or more specific rows from a dataset.


````python
#refer to the row index(to return row 0)
print(table.loc[0])
print(table.loc[2])
print("---------------------------------")
#To return row 0 and 1
print(table.loc[[0,1]])
print("---------------------------------")
df = pd.DataFrame(data, index = ["Person1", "Person2", "Person3"])
print(df)
print(df.loc[["Person1", "Person3"]])
print("---------------------------------")
print(df.loc["Person2"])
````

# **Data Cleaning**
Data cleaning is the process of correcting errors in your dataset, such as empty cells, incorrect formats, wrong information, and duplicate entries.

##**Cleaning Empty Cells**
Analyzing data with empty cells can lead to incorrect results.

###**Remove Rows**
- You can handle empty cells by deleting rows that have them. This is usually fine because datasets can be large, and getting rid of a few rows won't significantly affect the outcome.

````python
# Return new Data Frame with no empty cells if contains.

import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
new_df = df.dropna()
print(new_df.to_string())
# The dropna() method usually makes a new DataFrame and doesn't modify the original one by default.

# To modify the original DataFrame, use the 'inplace=True' option.
df.dropna(inplace=True)
print(df.to_string())
print(len(df))
````

## **Replace Empty Values**

- An alternative approach to handling empty cells is to replace them with a new value, eliminating the need to delete entire rows due to empty cells.
- The `fillna()` method enables us to substitute empty cells with a specified value.

## **Replace only for specified columns**
- Instead of replacing empty cells in the entire DataFrame, you can target a specific column by specifying its name.

##**Replace Using Mean, Median, or Mode**
- To fill empty cells, a common approach is to use the average (mean), middle value (median), or most frequent value (mode) of the column.
- In Pandas, you can achieve this by employing the `mean()`,` median()`, or `mode()` methods for the desired column, respectively.

###**Mean**
- Mean is the average, calculated by adding up all values and then dividing by the number of values.

###**Median**
- Median is the middle value when you arrange all the values in ascending order.

###**Mode**
 - Mode is the number that shows up the most.

````python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/test.csv")
df.fillna(55, inplace = True)
print(df.to_string())
print("------------------------------------------------------------------------------")
# Replacing the values in the specified column
df["LotFrontage"].fillna(35,inplace = True)
print("------------------------------------------------------------------------------")
# Replacing the values using Mean, Median and Mode
temp = df['LotFrontage'].mean()
print(temp)
df['LotFrontage'].fillna(temp,inplace = True)

temp1 = df['LotFrontage'].median()
print(temp1)
df['LotFrontage'].fillna(temp1,inplace = True)

temp2 = df['LotFrontage'].mode()
print(temp2)
df['LotFrontage'].fillna(temp2,inplace = True)
````

#**Data Correlations**

- The Pandas module offers a useful feature called the `corr()` method.
- This method computes the correlation between each column in your dataset, helping you understand the relationships between different data columns.

---------------------------------------------------------------
**Note:** The `corr()` method doesn't consider columns that are not numeric.



````python
df = pd.read_csv("/content/drive/MyDrive/data.csv")
df.corr()
````

###**Explanation**:

The `corr()` method produces a table (shown above) of numbers ranging from -1 to 1, indicating how closely two columns are related. A correlation of 1 implies a perfect positive relationship, where as one variable increases, the other also increases. A correlation of 0.9 suggests a strong positive relationship, meaning if one value rises, the other is likely to rise as well. Similarly, a correlation of -0.9 indicates a strong negative relationship, where an increase in one variable is associated with a decrease in the other. On the other hand, a correlation of 0.2 suggests a weak relationship, meaning that changes in one variable don't necessarily predict changes in the other.


##**Perfect Correlation:**
The "Duration" column has a correlation of 1.000000 with itself, which is expected because each column always perfectly correlates with itself.

##**Good Correlation**:
The relationship between "Duration" and "Calories" is strong, with a correlation of 0.922721. This means that if you exercise for a longer time, you're likely to burn more calories, and vice versa. In simpler terms, the longer you work out, the more calories you're likely to burn, and if you burned a lot of calories, it's likely you had a lengthy workout.
##**Bad Correlation**:

The relationship between "Duration" and "Maxpulse" is very weak, with a correlation of 0.009403. This suggests that there's not a reliable connection between the length of a workout and the maximum heart rate achieved. In simpler terms, we can't accurately predict someone's max heart rate just by knowing how long they worked out, and vice versa.

#**Plotting**

- Specify these axes using the x and y arguments in your code.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/diabetes.csv")
df.corr()
````

````python
df.plot()
plt.show()

print("------------------------------------------------------")
# Scatter Plot
df.plot(kind = "scatter", x = "BloodPressure", y = "Outcome")
plt.show()
df.plot(kind = "scatter", x = "BloodPressure", y = "DiabetesPedigreeFunction")
plt.show()
df.plot(kind = "scatter", x = "Pregnancies", y = "Age") # Good Correlation
plt.show()
df.plot(kind = "scatter", x = "DiabetesPedigreeFunction", y = "Age") # Bad Correlation
plt.show()
````
